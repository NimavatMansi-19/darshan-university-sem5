### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
a=data.columns
b=[]
for i in a:
    if(data[i].dtypes!=object):
        b.append(i)
for i in b:
    print(i,":")
    print("\tMean=",data[i].mean())
    print("\tStandard deviation=",data[i].std())
    print("\tMinimum=",data[i].min())
    print("\tMaximum=",data[i].max())
```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.

```python
plt.scatter(data['Age'],data["Fare"],alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs Fare")
plt.show()
```

```python
a=data.columns
b=[]
for i in a:
    if(data[i].dtypes!=object):
        b.append(i)
for i in range(len(b)-1):
    plt.scatter(data[b[i]],data[b[i+1]],alpha=0.5)
    plt.xlabel(b[i])
    plt.ylabel(b[i+1])
    plt.show()
#subplot
```
