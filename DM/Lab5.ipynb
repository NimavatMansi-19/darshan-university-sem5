### 2)  Handle Missing Values in data set [use dropna(), fillna(), and interpolate]


```python
dpna=dt.copy();
# dpna=dpna.dropna()
dpna=dpna.dropna(how='any',axis=1)
dpna
```

```python
filna=dt.copy()
# filna=filna.fillna(1000)
meanage=filna.Age.mean()
modecabin=filna.Cabin.mode()[0]
print(meanage)
print(modecabin)
filna=filna.fillna({'Age':meanage,'Cabin':modecabin})
filna
```

```python
inter=dt.copy()
inter=inter.interpolate()
inter
```

```python
a1=dt.Age.max()
a2=dt.Age.min()
dt['Age']=((dt['Age']-a2)/(a1-a2))
dt
```
